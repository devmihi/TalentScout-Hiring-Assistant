```python
print("ü§ñ TalentScout Hiring Assistant")
print("Hello! I will assist with initial candidate screening.")
print("Type 'exit' anytime to end the conversation.\n")
```

```python
questions = [
    "Full Name",
    "Email Address",
    "Phone Number",
    "Years of Experience",
    "Desired Position",
    "Current Location",
    "Tech Stack (languages, frameworks, tools)"
]

candidate_data = {}

for q in questions:
    answer = input(f"{q}: ")

    if answer.lower() == "exit":
        print("\nConversation ended by user.")
        break

    if not answer.strip():
        print("Please enter a valid response.")
        continue

    candidate_data[q] = answer
```

```python
print("\n‚úÖ Candidate Information Collected:")
for key, value in candidate_data.items():
    print(f"{key}: {value}")
```

```python
from google.colab import userdata
from groq import Groq

# Load Groq API key
client = Groq(api_key=userdata.get("GROQ_API_KEY"))

tech_stack = candidate_data.get("Tech Stack (languages, frameworks, tools)")

prompt = f"""
You are a technical interviewer at a recruitment agency.

The candidate has experience in the following tech stack:
{tech_stack}

Generate 3 to 5 technical interview questions that:
- Are suitable for initial screening
- Test practical understanding
- Vary in difficulty

Return the questions as a numbered list.
"""

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print("\nüîç Technical Screening Questions:")
print(response.choices[0].message.content)



```
